10. Найдите сумму всех элементов, сумму внутри строк, сумму внутри столбцов, а также среднее значение, дисперсию и стандартное отклонение чисел для каждой строки матрицы $A$. (Подобно тому, как sorted имеет необязательный аргумент key=, многие функции Numpy имеют необязательный аргумент axis=)


```python
A = np.array([[ 2.11741373,  2.98382182, -0.08017527,  0.53762474,  1.58569118],
       [-0.35738302,  2.18390944,  1.02698938, -0.32013668,  0.84642228],
       [ 2.57085885, -0.07038243,  2.1893022 ,  1.59759088,  2.1257642 ]])

print(A.sum())
print(A.sum(axis=1))  # По строкам
print(A.sum(axis=0))  # По столбцам
print(A.mean(axis=1))  # Срзнач
print(A.var(axis=1))  # У меня дисперсия
print(A.std(axis=1))  # Стандарт дивиейшн
```

13. Посчитайте две матрицы: $B = A^TA$ и $C = AA^T$. Что вы можете о них сказать? Какого они размера? В чём их сходство и отличие?

```python
A = np.random.uniform(-1, 3, 15).reshape(3, 5)
B = A.dot(A.T)
C = A.T.dot(A)
print('В =', B)
print('C =', C)

# Матрица В имеет размеры 3х3, матрица С - 5х5.
```
